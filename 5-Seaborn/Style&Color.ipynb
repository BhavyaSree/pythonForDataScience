```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x='sex', data=tips)
```

```python
# style argument sets the style for all the plots
sns.set_style('white')     # white background
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')     # ticks at the end of the box
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('darkgrid')     
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')     
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')     
sns.countplot(x='sex', data=tips)
sns.despine()     # to remove spines, by default it will remove right and top spines
```

```python
sns.set_style('ticks')     
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)     # we can specify the particular spines to remove
```

```python
# for figure size, we can use matplotlib
# through sns, we can use aspect ratio.
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

```python
# the figure to be a poster
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```

```python
# still need to change the size
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex', data=tips)
```

```python
# by default, the context will be notebook
sns.set_context('notebook')
sns.countplot(x='sex', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')

# different palette can be found in matplotlib colormap documentation
```
