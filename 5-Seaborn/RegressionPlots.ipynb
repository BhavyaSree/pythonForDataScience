```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex')

# to check male and female have same linear grid regarding to total_bill and tips
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'])

# different markers for different hue elements
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':100})

# scatter_kws, s element indicates the size of the markers
```

```python
# we can use a grid, instead of separating categories with hue
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex')

# instead of separating with colors, we separted with columns
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex', row='time')
```

```python
# to add more variables for plots, we can use hue
sns.lmplot(x='total_bill',y='tip',data=tips, col='day', hue='sex')
```

```python
# to adjust size and aspect ratio (ration of height and width of grid)

sns.lmplot(x='total_bill',y='tip',data=tips, col='day', hue='sex', aspect=0.6, size=8)
```

```python

sns.lmplot(x='total_bill',y='tip',data=tips, col='day', hue='sex', aspect=0.6, height=8)
```
