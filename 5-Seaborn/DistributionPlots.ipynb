```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')   # tips dataset available in seaborn
# https://github.com/mwaskom/seaborn-data
```

```python
# DIST plot - to show distribution of univariate(single variable) set of observations
sns.distplot(tips['total_bill'])    

# The curve is kde -- kernel density estimation
```

```python
sns.distplot(tips['total_bill'], kde=False)    # histogram is a distribution plot
```

```python
sns.distplot(tips['total_bill'], kde=False,bins=30)    
```

```python
# "bin" (or "bucket") the range of values—that is, divide the entire range of values into a series of intervals
# —and then count how many values fall into each interval
```

```python
# JOINT plot    -- to compare two variables
# we get two distribution plots for two variables and a scatter plot showing the relation between two variables

sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
# kind parameter effects the graph inside the joint plot,by default it will be scatter
# hex reprsents hexagon distribution representation.
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')

# with kind as reg, seaborn draws regression line on the scatter plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')

# with kind as kde, it shows the density where the points match up tme most
```

```python
# pairplot -- plot the pairwise relationships across the entire dataframe(numerical columns)

# pairplot do jointplot for every pair possible for numerical columns of dataset

sns.pairplot(tips)

# historm for x vs x
```

```python
# We can pass names of the categorical columns in hue argument.
# pairplot color the datapoints based on the data of hue column
sns.pairplot(tips, hue='sex')
```

```python
# palette is to choose the colors for the datapoints.
# Seaborn has some palette names with specific colors
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# KdePlots
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```
