```python
import seaborn as sns
%matplotlib inline 
# to see visualization in notbook
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
# barplot -- aggregate the categorical data based on mean
sns.barplot(x='sex',y='total_bill',data=tips)       # x-axis will be categorical, Y-axis will be numerical
```

```python
# the average total bill of male is higher than the average total bill of females
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)

# estimator argument represents the aggregate function to use to build barplot, by default, it will be mean
# we can give our own aggregator function too in estimator
```

```python
sns.countplot(x='sex',data=tips)      # number of occurences in the dataset,
# same as barplot, just Y-axis will be prechosen as count i.e., number of occurences

# This indicates more males in dataset than females
```

```python
sns.boxplot(x='day', y='total_bill', data=tips )

# x will be categorical, y will be numerical
# boxplot shows the distribution of data points

# distribution is separated by category of x variable
```

```python
# hue splits the boxplots in to further based on the categorical variable given

sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips )
```

```python
# violinplot   -- similar to boxplot, it shows density of the data points in the distribution

# it's hard to read the violinplot, though it gives more information than boxplot.

sns.violinplot(x='day', y='total_bill', data=tips )
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex' )
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True )
```

```python
# strip plot  -- to draw a scatter where one variable is categorical

sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)   

# jitter shows the density of data points little better
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex') 
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True) 
```

```python
# Violin and strip plots are hard to interpret
# boxplot is easy to interpret
```

```python
# swarmplot -- just like strip plot, only the points are adjusted so that they won't overlap

sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# swarm plot is the combination of strip and violin plots

# swarm plots won't be reliable for large datasets
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
# factorplot -- general form of all the plots

sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
# based on the kind, the catplot gives the graphs like bar,box,strip,swarm,violin
```
