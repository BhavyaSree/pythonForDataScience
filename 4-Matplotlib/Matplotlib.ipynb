```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# Two ways of creating Matplotlib plots
# 1. functional method
# 2. Object oriented method
```

### Functional method

```python
# Functional
plt.plot(x,y)
# plt.show()  -- to see the plot (if not in jupyter notebook)
```

```python
plt.plot(x,y,'r')     # graph in particular colour, here red
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# multiplots on the same canvas
plt.subplot(1,2,1)
# fisrt parameter - no.of rows
# second parameter - no.of.columns
# third parameter - plot number referring to
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

### Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])   # 1.left,2.bottom,3.width,4.height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()       # creating figure object, it is a empty canvas

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])   # here we manually add axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])  
# bottom of the set of axes starts 20% from the left, 50% from the bottom, 40% of the canvas width, 30% of the canvas height

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('SMALLER PLOT')
axes1.set_title('LARGER PLOT')
```

```python
fig = plt.figure()     # empty canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8])   # adding axes to the empty canvas
```

```python
# Subplots using object oriented method

fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)   # this automatically configure the axes based on rows and columns we want

plt.tight_layout()   # to fix overlapping layout
```

```python
axes     # have the list of the axes of subplots
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('FIRST PLOT')
axes[1].plot(y,x)
axes[1].set_title('SECOND PLOT')
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)

# figsize(3,2) indicates width and height of the figure in inches.
# dpi indicates dots per inch
```

```python
fig = plt.figure(figsize=(3,2))     # with default dpi

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))     # with default dpi

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

## Plot appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='green')    # accepts colors as strings for basic colors
ax.plot(x,y,color='#FF8C00')    # accepts RGB Hex codes, for custom color
```

```python
# To customize line width and line style
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='purple',linewidth=1)     # 1 indicates default linewidth, 2 indicates double the default linewidth
#ax.plot(x,y,color='purple',linewidth=3)
# ax.plot(x,y,color='purple',linewidth=3,alpha=0.5) 
# alpha is to control the traspency of the line
# ax.plot(x,y,color='purple',lw=3,alpha=0.5) 
ax.plot(x,y,color='purple',lw=3,linestyle='--') 
# differnt line styles like -,--,:,steps
```

```python
# to mark the points on the plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y,color='purple',lw=0.5,linestyle='--',marker='o')   # o is one the marker symbols avilable, some are +,*,1
# we can give marker size
ax.plot(x,y,color='purple',lw=0.5,linestyle='--',marker='o', markersize='10') 
```

```python
# we can also specify markerfacecolor(color isnide the circle)
# markeredgecolor (color of the marker edge)
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,linestyle='--',marker='o', 
        markersize='20', markerfacecolor='yellow', markeredgecolor='green')
```

### plot range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

# To show th plot between only 0 and 1 on x axis
ax.set_xlim([0,1])    # 1.lowerbound, 2. upperbound, in this case 0 is lowerbound and 1 is upperbound
ax.set_ylim([0,2])
```
